```python
%sql
SELECT
  HOUR(pickup_datetime)              AS hora_do_dia,
  ROUND(AVG(passenger_count), 2)     AS media_passageiros,
  COUNT(*)                           AS qtd_corridas
FROM ifood_case.silver.taxi_trips
WHERE pickup_datetime >= '2023-05-01'
  AND pickup_datetime <  '2023-06-01'
  AND passenger_count IS NOT NULL
  AND passenger_count > 0
GROUP BY 1
ORDER BY 1;
```

## Resposta
A média de passageiros por corrida em maio/2023 seguiu um padrão claro 
ao longo do dia: os valores mais altos ocorrem na madrugada (0h-3h) e 
à noite (20h-23h), em torno de 1,45 passageiros por corrida — coerente 
com deslocamentos em grupo após eventos sociais. O menor valor acontece 
por volta das 6h da manhã (~1,26), quando as corridas tendem a ser 
individuais (deslocamento solo para o trabalho). Ao longo do dia 
comercial (7h-19h), a média se mantém relativamente estável entre 1,35 
e 1,40.

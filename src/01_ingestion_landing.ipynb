```python
import os
import requests

BASE_URL = "https://d37ci6vzurychx.cloudfront.net/trip-data"
LANDING = "/Volumes/ifood_case/bronze/landing"
MONTHS = [f"2023-{m:02d}" for m in range(1, 6)]   # Jan a Mai/2023
TAXI_TYPES = ["yellow", "green"]

for taxi in TAXI_TYPES:
    os.makedirs(f"{LANDING}/{taxi}", exist_ok=True)
    for month in MONTHS:
        fname = f"{taxi}_tripdata_{month}.parquet"
        dest = f"{LANDING}/{taxi}/{fname}"
        if os.path.exists(dest):
            print(f"[skip] {fname} já existe")
            continue
        resp = requests.get(f"{BASE_URL}/{fname}", timeout=300)
        resp.raise_for_status()
        with open(dest, "wb") as f:
            f.write(resp.content)
        print(f"[ok] {fname} ({len(resp.content)/1e6:.1f} MB)")
```

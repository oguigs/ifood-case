```python
%sql
SELECT
  date_format(tpep_pickup_datetime, 'yyyy-MM') AS mes,
  ROUND(AVG(total_amount), 2)                  AS media_total_amount,
  COUNT(*)                                     AS qtd_corridas
FROM ifood_case.silver.yellow_taxi_trips
GROUP BY 1
ORDER BY 1;
```

```python
%sql
SELECT ROUND(AVG(receita_mensal), 2) AS receita_media_por_mes
FROM (
  SELECT date_format(tpep_pickup_datetime, 'yyyy-MM') AS mes,
         SUM(total_amount) AS receita_mensal
  FROM ifood_case.silver.yellow_taxi_trips
  GROUP BY 1
);
```
